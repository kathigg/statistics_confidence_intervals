```python
# Exponential distribution, lambda = 2
# E[x] = 0.5
# Var(x) = 0.25
# Std = sqrt(Var(x)) = 0.5
# F(t) = 1 - e^-2t
# t = -ln(1-u)/2
```

```python
import random
import numpy as np
from scipy import stats as st

n = 500
lambda_ = 2
alpha = 0.10 # 100(1-alpha)% = %90% ====> alpha = 0.10
sigma = 0.5
mu = 1/lambda_ # True mean

data = []
for i in range(n):
  u = random.random()
  t= -np.log(1-u)/2
  data.append(t)

z_score = st.norm.ppf(1-alpha/2)
xbar = np.mean(data)
lowerbound = xbar - z_score*sigma/np.sqrt(n)
upperbound = xbar + z_score*sigma/np.sqrt(n)
print("Confidence interval: (", lowerbound, ",", upperbound, ")")


```

```python
#1c
m = 100
count = 0

for i in range(m):
  data = []
  for i in range(n):
    u = random.random()
    t = -np.log(1-u)/2
    data.append(t)
  xbar = np.mean(data)
  lowerbound = xbar - z_score * sigma / np.sqrt(n)
  upperbound = xbar + z_score * sigma / np.sqrt(n)
  if lowerbound <= mu <= upperbound:
    count += 1
coverage_100 = count/m * 100

print("Confidence interval: ", coverage_100, "%")
```
